```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
columns=["sepal_width", "sepal_length", "species"]
df[columns]
```

```python
df["sepal_width"]
```

```python
df.sepal_width
```

```python
df.filter(regex='\.') #.이 들어간 것
df.filter(regex='_') #.이 들어간 것
```

```python
df.filter(regex='length$') #length로 끝나는 것
```

```python
df.filter(regex='^sepal') #Sepal로 시작하는 것
```

```python
df.filter(regex='^x[1-5]$') #x로 시작하고, 1에서 5로 끝나는 것
```

```python
df.filter(regex='es$')
```

```python
df.filter(regex='^(?!species$).*')
```
