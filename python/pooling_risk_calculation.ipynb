```python
import numpy as np
import pandas as pd
```

```python
from math import log10
 
def log_fact(n):
    return sum(log10(i) for i in range(1, n+1))

def nCr(n, r):
    a, b, c = log_fact(n), log_fact(r), log_fact(n-r)
    return 10**(a-b-c)
```

```python
prob_list=[]
for i in range(0, 50):
    prob = nCr(49, i)*((0.2)**i)*(0.8**(49-i))
    prob_list.append(prob)

print(prob_list)
```

```python
shared_loss_list =[0]
for i in range(1, 50):
    shared_loss= (i*2500)/49
    shared_loss_list.append(shared_loss)
shared_loss_list
```

```python
exp_list=[]
for i in range(0, 50):
    exp=shared_loss_list[i]*prob_list[i]
    exp_list.append(exp)    
```
