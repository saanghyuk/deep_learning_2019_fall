```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
s = pd.Series(np.random.randn(1000),     
              index=pd.date_range('1/1/2017', periods=1000))
```

```python
s
```

```python
s.plot()
```

```python
s=s.cumsum()
s.plot()
```

```python
s.shape
```

```python
r=s.rolling(window=60)
```

```python
r.mean() #이동평균 60일
```

```python
r.mean().isnull().sum()
```

```python
s.plot(style='k--')
r.mean().plot(style='k')
#그냥 그리면 추세선
```

```python
df = pd.DataFrame(np.random.randn(1000, 4),
                  index=pd.date_range('1/1/2017', periods=1000),
                  columns=['A', 'B', 'C', 'D'])
```

```python
df
```

```python
df=df.cumsum()
df
```

```python
df.plot()
```

```python
df.rolling(window=60).sum().plot(subplots=True)
#60일간의 합 추세선
```

```python
df
```

```python
len(df)
```

```python
df.rolling(window=len(df), min_periods=1).mean()[:5]
```

```python
df.rolling(window=len(df), min_periods=1).mean().plot()
```

```python
df.expanding(min_periods=1).mean().plot()
```

```python
df.expanding?
```

```python
dfe = pd.DataFrame({'B': [0, 1, 2, np.nan, 4]})
dfe
```

```python
dfe.plot()
```

```python
dfe.expanding(2).sum() #누적 합계
dfe.expanding(2).sum().plot()
```

```python
dfe.expanding(2).mean().plot()
```

expanding / rolling은 주로 이동평균 구할 때 사용함.
